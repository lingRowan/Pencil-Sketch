```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/Users/rowanahmed/Downloads/Three-Australian-Shepherd-puppies-sitting-in-a-field.jpg')
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
inverted_gray_image = cv2.bitwise_not(gray_image)
blurred_image = cv2.GaussianBlur(inverted_gray_image, (21, 21), sigmaX=0, sigmaY=0)
inverted_blurred_image = cv2.bitwise_not(blurred_image)
pencil_sketch_image = cv2.divide(gray_image, inverted_blurred_image, scale=256.0)
    
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  
plt.title("Original Image")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(pencil_sketch_image, cmap='gray')  
plt.title("Pencil Sketch")
plt.axis("off")
plt.show()
```
